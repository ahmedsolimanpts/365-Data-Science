```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url='https://en.wikipedia.org/wiki/List_of_largest_companies_by_revenue'
```

```python
response = requests.get(url) 
```

```python
html =response.content
```

```python
tables = pd.read_html(html,)
```

```python
company = tables[0]
```

```python
company.columns
```

```python
company.columns=['Rank','Name','Industry','Revenue','Profit','Employees','Headquarters','Ref']
company.index =company['Rank']
```

```python
company.drop(columns='Ref',inplace=True)
company
```
